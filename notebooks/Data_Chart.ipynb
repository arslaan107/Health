```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
pd.set_option('display.max_column', None)
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_csv("../data/graph1.csv")
df.head()
```

```python
df_renamed = df.copy()
```

```python
df_renamed.columns = [
    'Country',
    'Ctry_code',
    'Year',
    'Self_harm',
    'Forces_of_Nature',
    'Terrorism_conflict',
    'Interpersonal_violence',
    'Neglected_Tropical_Disease',
    'Substance_use',
    'Skin_diseases',
    'Enteric_infections',
    'Diabetes_kidney_diseases',
    'Cardiovascular_diseases',
    'Digestive_diseases',
    'Nutritional_deficiencies',
    'Respinf_and_TB',
    'Neonatal_disorders',
    'Chronic_respiratory_diseases',
    'Oth_non-communicable_diseases',
    'Maternal_disorders',
    'Unintentional_injuries',
    'Musculoskeletal_disorders',
    'Neoplasms',
    'Mental_disorders',
    'Neurological_disorders',
    'AIDS_STI',
    'Transport_injuries',
    'Sensory_organ_disease'
    
]
```

```python
df_renamed.head()
```

```python
df_ind = df_renamed.query("Country == 'India'")
```

```python
df_ind.head()
```

```python
df_ind['Year'] = df_ind['Year'].astype(str)
```

```python
df_ind.head()
```

```python
df_ind['Year'] = pd.to_datetime(df_ind['Year'])
```

```python
df_ind['Year'].head()
```

```python
df_ind['Year1'] = df_ind['Year'].dt.year
```

```python
df_ind['Year1'].head()
```

```python
import plotly.graph_objects as go
```

```python

fig = px.line(df_ind, x='Year', y = df_ind.columns[3:])

fig.show()
```

```python
df_ind = df_ind.drop("Year", axis='columns')
```

```python
df_ind.head()
```

```python
Year = df_ind['Year1']
df_ind.drop(labels=['Year1'], axis=1,inplace = True)
df_ind.insert(0, 'Year', Year)
```

```python
data = []
for col in df_ind.columns:
    data.append(go.Scatter(x = df_ind['Year'],
               y = df_ind[3:-20],
               mode = "lines",
               name = col))
    
    
layout = dict(
                width = 1000,
                height = 600,
                title = dict(text = "Question Count per Python Library", xanchor = "center", yanchor = 'top', x=0.45, y=0.9),
                xaxis = dict(title = "Date", range = [df_ind.index[0], df_ind.index[29]]),
                font = dict(size = 13),
                yaxis = dict(title = "Question Count", range = [0, 4500]),
                updatemenus = [dict(
                        type = "buttons",
                        buttons = [dict(
                                    label = "play",
                                    method = "animate",
                                    args = [None])])]
)

frames = [dict(data= [dict(type='scatter',
                           x = df_ind['Year'],
                           y = df_ind['Self_harm'][0:i]),
                      dict(type='scatter',
                           x = df_ind['Year'],
                           y = df_ind['Forces_of_Nature'][0:i]),
                      dict(type='scatter',
                           x = df_ind['Year'],
                           y = df_ind['Terrorism_conflict'][0:i]),
                      dict(type='scatter',
                           x = df_ind['Year'],
                           y = df_ind['Interpersonal_violence'][0:i]),
                      dict(type='scatter',
                           x = df_ind['Year'],
                           y = df_ind['Neglected_Tropical_Disease'][0:i])
                     ],
               traces= [0, 1, 2, 3],  
              )for i  in  range(1, len(df_ind)+1)]


fig = go.Figure(
    data = data,
    layout = layout,
    frames = frames
)



fig.show()
```
